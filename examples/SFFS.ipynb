```python
import physbo

import numpy as np

from mlxtend.feature_selection import SequentialFeatureSelector as SFS

# define target function

def target(x1,x2,x3,x4,x5):

    y = (100-x5)**2 + (2-x3)**2 + (10-x1)**2 + (1-x2)**2 + x4 + (1000-x1*x2) + np.random.normal(0,1)

    return y

para_num=2000

coefficient1 = np.random.uniform(-10,10,para_num)
coefficient2 = np.random.uniform(-10,10,para_num)
coefficient3 = np.random.uniform(-10,10,para_num)
coefficient4 = np.random.uniform(-10,10,para_num)
coefficient5 = np.random.uniform(-10,10,para_num)
coefficient = np.array([coefficient1,coefficient2,coefficient3,coefficient4,coefficient5]).T # coefficient = features X = [[coefficient1,coefficient3],[coefficient3,coefficient1]]

y = np.array([target(coefficient[i][0],coefficient[i][1],coefficient[i][2],coefficient[i][3],coefficient[i][4]) for i in range(para_num)])
```

```python
cov = physbo.gp.cov.gauss(coefficient.shape[1],ard = False)

mean = physbo.gp.mean.const()

lik = physbo.gp.lik.gauss()

config = physbo.misc.set_config()

gp = physbo.gp.sfs(lik=lik,mean=mean,cov=cov,config=config)

test = physbo.gp.model(lik=lik,mean=mean,cov=cov)
```

```python
sfs = SFS(estimator=gp,
          k_features=3,
          forward=True,
          floating=True,
          verbose=2,
          scoring='r2', 
          cv=3,clone_estimator=False,n_jobs=-1)
```

```python
sfs.fit(coefficient,y)
```

```python
sfs.subsets_ # CV score should converge to 1 # Test directly GP with 1 feature
```

```python
from mlxtend.plotting import plot_sequential_feature_selection as plot_sfs

fig = plot_sfs(sfs.get_metric_dict(), kind='std_err')
```
